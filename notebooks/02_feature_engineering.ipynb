# AI-Powered Financial Fraud Detection & Risk Analytics System

## Notebook 2: Data Preprocessing & Feature Engineering

### Objective

The objective of this notebook is to prepare the transaction dataset for machine learning by creating meaningful features, transforming raw attributes into analytical variables, and generating a model-ready dataset for fraud detection.

### Key Tasks

- Create transaction time features
- Create balance error features
- Generate customer and merchant attributes
- Create synthetic banking features
- Create behavioral and demographic attributes
- Prepare processed dataset for model training

### Input

../data/raw/paysim.csv

### Output

../data/processed/fraud_processed.csv

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/paysim.csv")

print(f"Dataset Shape: {df.shape}")

fraud_df = df.copy()

fraud_df.head()
```

```python
# Transaction Time Features

fraud_df["transaction_hour"] = fraud_df["step"] % 24

fraud_df["transaction_day"] = (
    fraud_df["step"] // 24
) + 1

fraud_df["transaction_week"] = (
    fraud_df["transaction_day"] // 7
) + 1

fraud_df[
    [
        "step",
        "transaction_hour",
        "transaction_day",
        "transaction_week"
    ]
].head()
```

```python
# Balance Error Features

fraud_df["balance_error_org"] = (
    fraud_df["oldbalanceOrg"]
    - fraud_df["newbalanceOrig"]
    - fraud_df["amount"]
)

fraud_df["balance_error_dest"] = (
    fraud_df["newbalanceDest"]
    - fraud_df["oldbalanceDest"]
    - fraud_df["amount"]
)

fraud_df[
    [
        "balance_error_org",
        "balance_error_dest"
    ]
].head()
```

```python
# Customer and Merchant Categories

fraud_df["customer_type"] = (
    fraud_df["nameOrig"].str[0]
)

fraud_df["merchant_type"] = (
    fraud_df["nameDest"].str[0]
)

fraud_df[
    [
        "customer_type",
        "merchant_type"
    ]
].head()
```

```python
# Age Group Feature

np.random.seed(42)

fraud_df["age_group"] = np.random.choice(
    [
        "18-25",
        "26-35",
        "36-45",
        "46-60",
        "60+"
    ],
    size=len(fraud_df),
    p=[
        0.20,
        0.35,
        0.25,
        0.15,
        0.05
    ]
)

fraud_df["age_group"].value_counts()
```

```python
# Customer Segment Feature

fraud_df["customer_segment"] = np.random.choice(
    [
        "Retail",
        "Premium",
        "Business"
    ],
    size=len(fraud_df),
    p=[
        0.70,
        0.20,
        0.10
    ]
)

fraud_df["customer_segment"].value_counts()
```

```python
# Banking Channel Feature

fraud_df["channel"] = np.random.choice(
    [
        "Mobile App",
        "Internet Banking",
        "ATM",
        "Branch"
    ],
    size=len(fraud_df),
    p=[
        0.50,
        0.25,
        0.15,
        0.10
    ]
)

fraud_df["channel"].value_counts()
```

```python
# Device Type Feature

fraud_df["device_type"] = np.random.choice(
    [
        "Android",
        "iPhone",
        "Desktop",
        "Tablet"
    ],
    size=len(fraud_df),
    p=[
        0.45,
        0.25,
        0.20,
        0.10
    ]
)

fraud_df["device_type"].value_counts()
```

```python
# Region Feature

fraud_df["risk_region"] = np.random.choice(
    [
        "North",
        "South",
        "East",
        "West",
        "Central"
    ],
    size=len(fraud_df)
)

fraud_df["risk_region"].value_counts()
```

```python
# Account Tenure Feature

fraud_df["account_tenure_months"] = np.random.randint(
    1,
    120,
    size=len(fraud_df)
)

fraud_df["account_tenure_months"].describe()
```

```python
print(f"Final Dataset Shape: {fraud_df.shape}")

fraud_df.head()
```

```python
fraud_df.to_csv(
    "../data/processed/fraud_processed.csv",
    index=False
)

print("Processed Dataset Saved Successfully")
```
