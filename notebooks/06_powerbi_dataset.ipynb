```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(
    "../data/processed/fraud_risk_scored.csv"
)

print(df.shape)
```

```python
total_transactions = len(df)

print(total_transactions)
```

```python
total_frauds = df["isFraud"].sum()

print(total_frauds)
```

```python
fraud_rate = (
    total_frauds /
    total_transactions
) * 100

print(round(fraud_rate,4))
```

```python
total_amount = df["amount"].sum()

print(total_amount)
```

```python
avg_amount = df["amount"].mean()

print(avg_amount)
```

```python
critical_alerts = (
    df["risk_category"] == "Critical"
).sum()

print(critical_alerts)
```

```python
high_risk = (
    df["risk_category"] == "High"
).sum()

print(high_risk)
```

```python
investigation_required = (
    df["investigation_required"] == "Yes"
).sum()

print(investigation_required)
```

```python
review_required = (
    df["review_required"] == "Yes"
).sum()

print(review_required)
```

```python
executive_kpis = pd.DataFrame({

    "KPI": [

        "Total Transactions",
        "Fraud Transactions",
        "Fraud Rate (%)",
        "Total Transaction Amount",
        "Average Transaction Amount",
        "Critical Alerts",
        "High Risk Transactions",
        "Investigation Required",
        "Review Required"

    ],

    "Value": [

        total_transactions,
        total_frauds,
        round(fraud_rate,4),
        round(total_amount,2),
        round(avg_amount,2),
        critical_alerts,
        high_risk,
        investigation_required,
        review_required

    ]

})

executive_kpis
```

```python
executive_kpis.to_csv(
    "../data/processed/executive_kpis.csv",
    index=False
)

print("Executive KPI Dataset Saved")
```

```python
executive_kpis
```

```python
risk_summary = (

    df.groupby(
        "risk_category"
    )

    .agg(

        transaction_count = (
            "risk_category",
            "count"
        ),

        avg_risk_score = (
            "risk_score",
            "mean"
        ),

        total_amount = (
            "amount",
            "sum"
        )

    )

    .reset_index()

)

risk_summary
```

```python
risk_order = {

    "Low":1,
    "Medium":2,
    "High":3,
    "Critical":4

}

risk_summary["sort_order"] = (

    risk_summary["risk_category"]

    .map(risk_order)

)

risk_summary = (

    risk_summary

    .sort_values(
        "sort_order"
    )

)

risk_summary
```

```python
risk_summary[
    "percentage"
] = (

    risk_summary[
        "transaction_count"
    ]

    /

    risk_summary[
        "transaction_count"
    ].sum()

    * 100

).round(4)

risk_summary
```

```python
risk_summary["percentage"] = (
    risk_summary["percentage"]
    .round(2)
)
```

```python
risk_summary.to_csv(
    "../data/processed/risk_summary.csv",
    index=False
)

print(
    "Risk Summary Saved Successfully"
)
```

```python
risk_summary
```

```python
severity_summary = (

    df.groupby(
        "fraud_severity"
    )

    .agg(

        transaction_count=(
            "fraud_severity",
            "count"
        ),

        avg_risk_score=(
            "risk_score",
            "mean"
        ),

        total_amount=(
            "amount",
            "sum"
        )

    )

    .reset_index()

)

severity_order = {

    "No Risk":1,
    "Minor":2,
    "Moderate":3,
    "Severe":4,
    "Extreme":5

}

severity_summary["sort_order"] = (
    severity_summary["fraud_severity"]
    .map(severity_order)
)

severity_summary = (
    severity_summary
    .sort_values("sort_order")
)

severity_summary["percentage"] = (

    severity_summary["transaction_count"]

    /

    severity_summary["transaction_count"].sum()

    * 100

).round(4)

severity_summary
```
