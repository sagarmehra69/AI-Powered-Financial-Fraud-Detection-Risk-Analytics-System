```python
# Import Libraries

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
# Load Risk Scored Dataset

df = pd.read_csv(
    "../data/processed/fraud_risk_scored.csv"
)

print(f"Dataset Shape: {df.shape}")

df.head()
```

```python
# Generate Executive KPI Dataset

total_transactions = len(df)

total_frauds = df["isFraud"].sum()

fraud_rate = (
    total_frauds
    / total_transactions
    * 100
)

total_amount = df["amount"].sum()

avg_amount = df["amount"].mean()

critical_alerts = (
    df["risk_category"] == "Critical"
).sum()

high_risk_transactions = (
    df["risk_category"] == "High"
).sum()

investigation_required = (
    df["investigation_required"] == "Yes"
).sum()

review_required = (
    df["review_required"] == "Yes"
).sum()

executive_kpis = pd.DataFrame({

    "KPI":[

        "Total Transactions",
        "Fraud Transactions",
        "Fraud Rate (%)",
        "Total Transaction Amount",
        "Average Transaction Amount",
        "Critical Alerts",
        "High Risk Transactions",
        "Investigation Required",
        "Review Required"

    ],

    "Value":[

        total_transactions,
        total_frauds,
        round(fraud_rate,4),
        round(total_amount,2),
        round(avg_amount,2),
        critical_alerts,
        high_risk_transactions,
        investigation_required,
        review_required

    ]

})

executive_kpis

executive_kpis.to_csv(
    "../data/processed/executive_kpis.csv",
    index=False
)

print("Executive KPI Dataset Saved Successfully")
```

```python
# Generate Risk Category Summary

risk_summary = (

    df.groupby(
        "risk_category"
    )

    .agg(

        transaction_count=(
            "risk_category",
            "count"
        ),

        avg_risk_score=(
            "risk_score",
            "mean"
        ),

        total_amount=(
            "amount",
            "sum"
        )

    )

    .reset_index()

)

risk_order = {

    "Low":1,
    "Medium":2,
    "High":3,
    "Critical":4

}

risk_summary["sort_order"] = (
    risk_summary["risk_category"]
    .map(risk_order)
)

risk_summary = (
    risk_summary
    .sort_values("sort_order")
)

risk_summary["percentage"] = (

    risk_summary["transaction_count"]

    /

    risk_summary["transaction_count"].sum()

    * 100

).round(2)

risk_summary

risk_summary.to_csv(
    "../data/processed/risk_summary.csv",
    index=False
)

print("Risk Summary Saved Successfully")
```

```python
# Generate Fraud Severity Summary

severity_summary = (

    df.groupby(
        "fraud_severity"
    )

    .agg(

        transaction_count=(
            "fraud_severity",
            "count"
        ),

        avg_risk_score=(
            "risk_score",
            "mean"
        ),

        total_amount=(
            "amount",
            "sum"
        )

    )

    .reset_index()

)

severity_order = {

    "No Risk":1,
    "Minor":2,
    "Moderate":3,
    "Severe":4,
    "Extreme":5

}

severity_summary["sort_order"] = (
    severity_summary["fraud_severity"]
    .map(severity_order)
)

severity_summary = (
    severity_summary
    .sort_values("sort_order")
)

severity_summary["percentage"] = (

    severity_summary["transaction_count"]

    /

    severity_summary["transaction_count"].sum()

    * 100

).round(2)

severity_summary

severity_summary.to_csv(
    "../data/processed/severity_summary.csv",
    index=False
)

print("Severity Summary Saved Successfully")
```

```python
# Generate Regional Risk Analysis

regional_summary = (

    df.groupby(
        "risk_region"
    )

    .agg(

        total_transactions=(
            "risk_region",
            "count"
        ),

        fraud_transactions=(
            "isFraud",
            "sum"
        ),

        avg_risk_score=(
            "risk_score",
            "mean"
        ),

        total_amount=(
            "amount",
            "sum"
        )

    )

    .reset_index()

)

regional_summary["fraud_rate"] = (

    regional_summary["fraud_transactions"]

    /

    regional_summary["total_transactions"]

    * 100

).round(4)

regional_summary = (
    regional_summary
    .sort_values(
        "fraud_rate",
        ascending=False
    )
)

regional_summary

regional_summary.to_csv(
    "../data/processed/regional_summary.csv",
    index=False
)

print("Regional Summary Saved Successfully")
```

```python
# Generate Customer Segment Analytics

customer_summary = (

    df.groupby(
        "customer_segment"
    )

    .agg(

        total_transactions=(
            "customer_segment",
            "count"
        ),

        fraud_transactions=(
            "isFraud",
            "sum"
        ),

        avg_risk_score=(
            "risk_score",
            "mean"
        ),

        avg_transaction_amount=(
            "amount",
            "mean"
        )

    )

    .reset_index()

)

customer_summary["fraud_rate"] = (

    customer_summary["fraud_transactions"]

    /

    customer_summary["total_transactions"]

    * 100

).round(4)

customer_summary = (
    customer_summary
    .sort_values(
        "fraud_rate",
        ascending=False
    )
)

customer_summary

customer_summary.to_csv(
    "../data/processed/customer_summary.csv",
    index=False
)

print("Customer Summary Saved Successfully")
```

```python
# Generate Channel Analytics

channel_summary = (

    df.groupby(
        "channel"
    )

    .agg(

        total_transactions=(
            "channel",
            "count"
        ),

        fraud_transactions=(
            "isFraud",
            "sum"
        ),

        avg_risk_score=(
            "risk_score",
            "mean"
        ),

        total_amount=(
            "amount",
            "sum"
        )

    )

    .reset_index()

)

channel_summary["fraud_rate"] = (

    channel_summary["fraud_transactions"]

    /

    channel_summary["total_transactions"]

    * 100

).round(4)

channel_summary = (
    channel_summary
    .sort_values(
        "fraud_rate",
        ascending=False
    )
)

channel_summary

channel_summary.to_csv(
    "../data/processed/channel_summary.csv",
    index=False
)

print("Channel Summary Saved Successfully")
```

```python
# Generate Device Analytics

device_summary = (

    df.groupby(
        "device_type"
    )

    .agg(

        total_transactions=(
            "device_type",
            "count"
        ),

        fraud_transactions=(
            "isFraud",
            "sum"
        ),

        avg_risk_score=(
            "risk_score",
            "mean"
        )

    )

    .reset_index()

)

device_summary["fraud_rate"] = (

    device_summary["fraud_transactions"]

    /

    device_summary["total_transactions"]

    * 100

).round(4)

device_summary

device_summary.to_csv(
    "../data/processed/device_summary.csv",
    index=False
)

print("Device Summary Saved Successfully")
```

```python
# Generate Age Group Analytics

age_summary = (

    df.groupby(
        "age_group"
    )

    .agg(

        total_transactions=(
            "age_group",
            "count"
        ),

        fraud_transactions=(
            "isFraud",
            "sum"
        ),

        avg_risk_score=(
            "risk_score",
            "mean"
        )

    )

    .reset_index()

)

age_summary["fraud_rate"] = (

    age_summary["fraud_transactions"]

    /

    age_summary["total_transactions"]

    * 100

).round(4)

age_summary

age_summary.to_csv(
    "../data/processed/age_summary.csv",
    index=False
)

print("Age Summary Saved Successfully")
```

```python
# ============================================================
# Create Power BI Master Dataset
# ============================================================

# Create simulation date
start_date = pd.Timestamp("2025-01-01")

# Create DateTime from step (step = hour)
df["transaction_datetime"] = (
    start_date + pd.to_timedelta(df["step"], unit="h")
)

# Date Dimensions
df["Date"] = df["transaction_datetime"].dt.date
df["Year"] = df["transaction_datetime"].dt.year
df["Quarter"] = "Q" + df["transaction_datetime"].dt.quarter.astype(str)
df["Month"] = df["transaction_datetime"].dt.month_name()
df["Month_Number"] = df["transaction_datetime"].dt.month
df["Week"] = df["transaction_datetime"].dt.isocalendar().week.astype(int)
df["Day"] = df["transaction_datetime"].dt.day_name()
df["Day_Number"] = df["transaction_datetime"].dt.dayofweek + 1
df["Hour"] = df["transaction_datetime"].dt.hour

# Create Power BI Master Dataset
powerbi_master = df[
[
    # Date & Time
    "Date",
    "Year",
    "Quarter",
    "Month",
    "Month_Number",
    "Week",
    "Day",
    "Day_Number",
    "Hour",

    # Transaction Information
    "step",
    "type",
    "amount",

    # Balance Information
    "oldbalanceOrg",
    "newbalanceOrig",
    "oldbalanceDest",
    "newbalanceDest",

    # Fraud Labels
    "isFraud",
    "isFlaggedFraud",

    # Engineered Features
    "balance_error_org",
    "balance_error_dest",

    # Customer Information
    "customer_type",
    "customer_segment",
    "age_group",
    "account_tenure_months",

    # Merchant Information
    "merchant_type",

    # Channel & Device
    "channel",
    "device_type",

    # Region
    "risk_region",

    # AI Risk Outputs
    "fraud_probability",
    "risk_score",
    "risk_category",
    "fraud_severity",

    # Investigation
    "alert_priority",
    "recommended_action",
    "alert_message",
    "review_required",
    "investigation_required"
]
]

print("Power BI Master Dataset Shape:", powerbi_master.shape)

powerbi_master.head()
```

```python
# Export Power BI Master Dataset

powerbi_master.to_csv(
    "../data/processed/powerbi_master_dataset.csv",
    index=False
)

print("Power BI Master Dataset Saved Successfully")
```
